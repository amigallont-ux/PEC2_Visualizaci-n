```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
import geopandas as gpd
import contextily as ctx

# Cargar datos
df = pd.read_csv("all_month.csv")

# Filtrar Europa (importantísimo)
df = df[(df["longitude"] > -20) & (df["longitude"] < 40) &
        (df["latitude"] > 30) & (df["latitude"] < 70)]

df = df[df["mag"] > 4]

# Crear GeoDataFrame
gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df.longitude, df.latitude),
    crs="EPSG:4326"
)

# Convertir a sistema web (para mapa)
gdf = gdf.to_crs(epsg=3857)

# Obtener puntos
points = gdf.geometry.apply(lambda p: [p.x, p.y]).to_list()

import numpy as np
points = np.array(points)

# Convex hull
hull = ConvexHull(points)

# Plot
fig, ax = plt.subplots(figsize=(10, 8))

# puntos
ax.scatter(points[:,0], points[:,1], s=10)

# hull
for simplex in hull.simplices:
    ax.plot(points[simplex,0], points[simplex,1], 'r-')

# mapa de fondo
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

ax.set_title("Convex Hull of Earthquakes (Europe)")
ax.set_axis_off()
ax.set_xlim(-1.5e6, 5e6)
ax.set_ylim(3.5e6, 7e6)
ax.set_title("Convex Hull of Earthquakes (Europe)")
ax.set_axis_off()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
import geopandas as gpd
import contextily as ctx

# 1. Cargar datos
df = pd.read_csv("all_month.csv")

# 2. Filtrar una zona amplia de Europa
#    Longitud: Portugal -> Hungría aprox. (-12 a 25)
#    Latitud:  España -> Noruega aprox. (35 a 72)
df_europe = df[
    (df["longitude"] >= -12) & (df["longitude"] <= 25) &
    (df["latitude"] >= 35) & (df["latitude"] <= 72)
].copy()

# 3. Eliminar nulos en coordenadas
df_europe = df_europe.dropna(subset=["longitude", "latitude"])

# 4. Crear GeoDataFrame
gdf = gpd.GeoDataFrame(
    df_europe,
    geometry=gpd.points_from_xy(df_europe["longitude"], df_europe["latitude"]),
    crs="EPSG:4326"
)

# 5. Pasar a Web Mercator para usar contextily
gdf = gdf.to_crs(epsg=3857)

# 6. Extraer coordenadas
points = np.array([[geom.x, geom.y] for geom in gdf.geometry])

# 7. Comprobar que hay suficientes puntos
if len(points) < 3:
    raise ValueError("No hay suficientes puntos para calcular la convex hull.")

# 8. Calcular convex hull
hull = ConvexHull(points)

# 9. Crear figura
fig, ax = plt.subplots(figsize=(12, 10))

# 10. Dibujar puntos
ax.scatter(points[:, 0], points[:, 1], s=18, alpha=0.7)

# 11. Dibujar aristas de la convex hull
for simplex in hull.simplices:
    ax.plot(points[simplex, 0], points[simplex, 1], linewidth=2)

# 12. Añadir mapa base
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

# 13. Fijar extensión para ver Europa entera en el rango pedido
#     Aproximadamente: Portugal -> Hungría y España -> Noruega
west, south = gpd.GeoSeries(
    gpd.points_from_xy([-12], [35]), crs="EPSG:4326"
).to_crs(epsg=3857).geometry.iloc[0].coords[0]

east, north = gpd.GeoSeries(
    gpd.points_from_xy([25], [72]), crs="EPSG:4326"
).to_crs(epsg=3857).geometry.iloc[0].coords[0]

ax.set_xlim(west, east)
ax.set_ylim(south, north)

# 14. Título y formato
ax.set_title("Convex Hull of Earthquakes in Europe", fontsize=14)
ax.set_axis_off()

plt.tight_layout()
plt.show()
```
